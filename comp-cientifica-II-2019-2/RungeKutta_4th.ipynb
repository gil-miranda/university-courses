```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def rk_4(F, y0, ts, p = 0):
    h = ts[1] - ts[0]
    ys = np.array([y0])
    for tnext in ts[1:]:
        k1 = F(tnext, ys[-1])
        k2 = F(tnext + h/2, ys[-1]+h*k1/2)
        k3 = F(tnext + h/2, ys[-1] + h*k2/2)
        k4 = F(tnext + h, ys[-1] + h*k3)
        ynext = ys[-1] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
        ys = np.vstack((ys, ynext))
    return ys
```

```python
def F(t, u):
    return -5 * u

def solution(t):
    return np.exp(-5*t)

u_0 = 1
hs = [0.1,0.05,0.025,0.0125]
color = ['red', 'darkblue', 'green', 'purple']
err = []
```

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
for ax,h,c in zip([ax1, ax2, ax3, ax4],hs, color):
    ts = np.arange(0,2,h)
    sol = rk_4(F, 1, ts)
    err.append(abs(sol - solution(ts)))
    ax.plot(ts, sol, color = c)
    ax.set_title('$ h = ' + str(h)+ '$')
    ax.grid(alpha = 0.3)
plt.suptitle('Runge Kutta com passo variando', fontweight = 'bold')
plt.show()
    
```
