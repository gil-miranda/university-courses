Queremos resolver:

$$ -\nabla(\kappa \nabla T) = 0 $$

A discretização será:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n+1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n+N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$

$$\begin{cases} C_{u_{n+1}} = \frac{\kappa\left(x_{i+\frac{1}{2},j}\right)}{h_x^2} \\ C_{u_n} = -\frac{\kappa\left(x_{i+\frac{1}{2},j}\right)}{h_x^2}-\frac{\kappa\left(x_{i-\frac{1}{2},j}\right)}{h_x^2} - \frac{\kappa\left(x_{i,j+\frac{1}{2}}\right)}{h_y^2} -\frac{\kappa\left(x_{i,j-\frac{1}{2}}\right)}{h_y^2} \\ C_{u_{n-1}} = \frac{\kappa\left(x_{i-\frac{1}{2},j}\right)}{h_x^2} \\ C_{u_{n+N_x}} = \frac{\kappa\left(x_{i,j+\frac{1}{2}}\right)}{h_y^2} \\
C_{u_{n-N_x}} = \frac{\kappa\left(x_{i,j-\frac{1}{2}}\right)}{h_y^2}
\end{cases}$$

### Nas condições de contorno, para i = 1 vamos usar a condiçao de Newton, que pode ser escrita como:

$$-\kappa(x_i,y_j)\frac{T_{i+1,j} - T_{i-1,j}}{2h_x} = 25(T_{i,j} + 11) $$

$$T_{i+1,j}= \frac{-50 h_x(T_{i,j} + 11)}{\kappa(x_i,y_j)} + T_{i-1,j} $$

$$T_{n-1}= \frac{50 h_x(T_{n} + 11)}{\kappa(x_i,y_j)} + T_{n+1} $$


Desta forma, o que precisamos agora é substituir isto na discretização e descobrir os novos coeficientes:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n+1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-\frac{50 h_x(T_{n} + 11)}{\kappa(x_i,y_j)} - T_{n+1} }{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n+N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$

O novo coeficiente de $T_n$ e $T_{n+1}$:

$$ C(T_n) = C(T_n) + \frac{50 \kappa_{left}}{h_x \kappa_{point}}$$

$$ C(T_{n+1}) = C(T_{n+1}) + \frac{\kappa_{left}}{h_x^2} $$

E devemos somar algo ao vetor b:

$$ \frac{550 \kappa_{left}}{h_x\kappa_{point}} $$

### Para j = 1, teremos:

$$-\kappa(x_i,y_j)\frac{T_{i,j+1} - T_{i,j-1}}{2h_y} = 25(T_{i,j} + 11) $$

$$T_{i,j+1}= \frac{-50 h_y(T_{i,j} + 11)}{\kappa(x_i,y_j)} + T_{i,j-1} $$

$$T_{n-N_x}= \frac{50 h_y(T_{n} + 11)}{\kappa(x_i,y_j)} + T_{n+Nx} $$


Substituindo na discretização:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n+1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n+N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-\frac{50 h_y(T_{n} + 11)}{\kappa(x_i,y_j)} - T_{n+Nx}}{h_y}\right) = 0$$

O novo coeficiente de $T_n$ e $T_{n+N_x}$:

$$ C(T_n) = C(T_n) + \frac{50 \kappa_{down}}{h_y \kappa_{point}}$$

$$ C(T_{n+N_x}) = C(T_{n+N_x}) + \frac{\kappa_{down}}{h_y^2} $$

Devemos somar algo ao vetor b:

$$ \frac{550 \kappa_{down}}{h_y\kappa_{point}} $$

### Para $i = N_x$ e ys < 0.61:

$$ -\kappa(x_i,y_j)\frac{T_{i+1,j}-T_{i-1,j}}{h_x} = 0 $$

$$ T_{i,j} = T_{i-1,j} $$

$$ T_{n+1} = T_{n-1} $$


Substituindo na discretização:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n-1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n+N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$

O novo coeficiente de $T_{n-1}$ será:

$$C(T_{n-1}) = C(T_{n-1}) + \frac{\kappa(x_{i+1/2,j})}{h_x^2} $$

### Para $j = N_y$ e x < 0.61:

$$ -\kappa(x_i,y_j)\frac{T_{i,j+1}-T_{i,j-1}}{2h_y} = 0 $$

$$ T_{i,j+1} = T_{i,j-1} $$

$$ T_{n+N_x} = T_{n-N_x} $$

Substituindo na discretização:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n+1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n-N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$

O novo coeficiente de $T_{n-N_x}$ será:

$$C(T_{n-N_x}) = C(T_{n-N_x}) + \frac{\kappa_{up}}{h_y^2} $$

### Se xs ou ys > 0.61:

### Se $xs[i] = 0.61$ e $ys[j] > 0.61 $, então teremos condição de contorno em $i-1$ da forma:

$$-\kappa(x_i,y_j)\frac{T_{i+1,j} - T_{i-1,j}}{2h_x} = 6(T_{i,j} - 20) $$

$$T_{i+1,j}= \frac{-12 h_x(T_{i,j} - 20)}{\kappa(x_i,y_j)} + T_{i-1,j} $$

$$T_{n+1}= \frac{-12 h_x(T_{n} - 20)}{\kappa(x_i,y_j)} + T_{n-1} $$


Substituindo na discretização:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{\frac{-12 h_x(T_{n} - 20)}{\kappa(x_i,y_j)} + T_{n-1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{T_{n+N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$

O novo coeficiente de $T_n$ e $T_{n-1}$:

$$ C(T_n) = C(T_n) + \frac{-12 \kappa_{left}}{h_x \kappa_{point}}$$

$$ C(T_{n-1}) = C(T_{n-1}) + \frac{\kappa_{right}}{h_x^2} $$


E devemos somar algo ao vetor b:

$$ \frac{240 \kappa_{left}}{h_x\kappa_{point}} $$

### Se $xs[i] > 0.61$ e $ys[j] = 0.61 $, então teremos condição de contorno em $j-1$ da forma:

$$-\kappa(x_i,y_j)\frac{T_{i,j+1} - T_{i,j-1}}{2h_y} = 6(T_{i,j} - 20) $$

$$T_{i,j+1}= \frac{-12 h_y(T_{i,j} - 20)}{\kappa(x_i,y_j)} + T_{i,j-1} $$

$$T_{n+N_x}= \frac{-12 h_y(T_{n} - 20)}{\kappa(x_i,y_j)} + T_{n-N_x} $$

Substituindo na discretização:

$$ \frac{1}{h_x}\left(\kappa(x_{i+\frac{1}{2},j})\frac{T_{n+1}-T_{n}}{h_x}-\kappa(x_{i-\frac{1}{2},j})\frac{T_{n}-T_{n-1}}{h_x}\right) + \frac{1}{h_y}\left(\kappa(x_{i,j+\frac{1}{2}})\frac{\frac{-12 h_y(T_{n} - 20)}{\kappa(x_i,y_j)} + T_{n-N_x}-T_{n}}{h_y}-\kappa(x_{i,j-\frac{1}{2}})\frac{T_{n}-T_{n-N_x}}{h_y}\right) = 0$$


O novo coeficiente de $T_n$ e $T_{n-N_x}$:

$$ C(T_n) = C(T_n) + \frac{-12 \kappa_{up}}{h_y \kappa_{point}}$$

$$ C(T_{n-N_x}) = C(T_{n-N_x}) + \frac{\kappa_{up}}{h_y^2} $$


E devemos somar algo ao vetor b:

$$ \frac{240 \kappa_{up}}{h_y\kappa_{point}} $$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
def kappa(x,y):
    if (x<=0.02 or y<=0.02):
        return 0.81
    elif (0.02<=x<=1.61 and 0.02<=y<=0.14) or (0.02<=x<=0.14 and 0.02<=y<=1.61):
        return 0.049
    elif (0.14<=x<=1.61 and 0.14<=y<=0.59) or (0.14<=x<=0.59 and 0.14<=y<=1.61):
        return 0.5
    elif (0.59<=x<=1.61 and 0.59<=y<=0.61) or (0.59<=x<=0.61 and 0.59<=y<=1.61):
        return 0.99

def L(x,y):
    if (0 <= x <= 0.61 and 0 <= y <= 1.61) or  (0 <= x <= 1.61 and 0 <= y <= 0.61):
        return True
    else:
        return False
    
def heat_equation(kappa,Nx,Ny):
    A = []
    b = np.zeros(Nx*Ny)
    hx = 1.61/(Nx-1)
    hy = 1.61/(Ny-1)
    xs = [(i-1)*hx for i in range(1,Nx+1)]
    ys = [(j-1)*hy for j in range(1,Ny+1)]
    for n in range(Nx*Ny):
        j = n//Nx+1
        i = n - Nx*(j-1)+1
        xi = xs[i-1]
        yj = ys[j-1]

        
        test = L(xi,yj)
        
        xm = xi-hx/2
        xM = xi + hx/2
        ym = yj - hy/2
        yM = yj + hy/2
        
        k_p = kappa(xi,yj)
        k_l = kappa(xm,yj)
        k_r = kappa(xM,yj)
        k_u = kappa(xi,yM)
        k_d = kappa(xi,ym)
        
        a = np.zeros(Nx*Ny)
        if test == False:
            a[n] = 1
            b[n] = 0
        else:
            if i == Nx:
                k_r = k_p
                a[n-1] += k_r/hx**2
            elif k_r != None:
                a[n+1] += k_r/hx**2
                
            if np.isclose(xi,0.61,atol = hx) and yj > 0.61:
                k_r = k_p
                a[n] -= 12/hx
                a[n-1] += k_r/hx**2
                b[n] -= 240/hx
                
        
            if j == Ny:
                k_u = k_p
                a[n-Nx] += k_u/hy**2
            elif k_u != None:
                a[n+Nx] += k_u/hy**2
            if np.isclose(yj,0.61,atol = hy) and xi > 0.61:
                k_u = k_p
                a[n] -= 12/hy
                a[n-Nx] += k_u/hy**2
                b[n] -= 240/hy
            if i == 1:
                k_l = k_p
                a[n] += 50/hx
                a[n+1] += k_l/hx**2
                b[n] -= 550/hx
            elif k_l != None:
                a[n-1] += k_l/hx**2
            if j == 1:
                k_d = k_p
                a[n] += 50/hy
                a[n+Nx] += k_d/hy**2
                b[n] -= 550/hy
            elif k_d != None:
                a[n-Nx] += k_d/hy**2
            a[n] += -k_l/hx**2 -k_r/hx**2 - k_u/hy**2 - k_d/hy**2

        A.append(a)
    ys = np.linalg.solve(A,b)
    M = []
    for i in range(0,len(ys),Nx):
        y = ys[i:i+Nx]
        M.append(y)

    return np.array(M)

ys = heat_equation(kappa,100,100)
```

```python
plt.figure(figsize=(10,10))
plt.title("Heat Equation in a L dominion")
plt.contourf(ys,cmap='inferno')
plt.show()
plt.title("Heat map")
sn.heatmap(ys,vmin=-11,vmax=20)
plt.show()
```
