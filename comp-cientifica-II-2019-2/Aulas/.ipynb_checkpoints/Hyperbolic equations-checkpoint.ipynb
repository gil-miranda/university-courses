Hyperbollic Equations

$$
\frac{\partial U}{\partial t} + a \frac{\partial U}{\partial x} = 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nx = 10
Nt = 5
hx = 1/(Nx+1)
ht = 0.01
tf = 2
a = 1
sigma = a*ht/hx

## Boundary conditions
left_x = 0
right_x = 1


xs = np.arange(0,1+hx,hx)

ts = np.arange(0,tf+ht, ht)

Us = [[]]

## setting initial condition on t = 0
for x in xs:
    if (x < 0.5):
        newU = 0
    else:
        newU = 1
    Us[-1].append(newU)
        
        
## solving for t+1

for t in ts[:-1]:
    U = []
    for x in range(1,len(xs)-1):   
        newU = Us[-1][x] - sigma/2 * (Us[-1][x+1] - Us[-1][x-1])
        U.append(newU)
    U.append(1)
    U.insert(0,0)
    Us.append(U)
```

```python
plt.figure(figsize=(12,10))
for i in range(len(Us)):
    plt.plot(xs,Us[i])
plt.grid(alpha = 0.5)
plt.show()
```

```python
Nx = 20
Nt = 30
hx = 1/(Nx+1)
ht = 0.01
tf = 2
a = 1
sigma = a*ht/hx

## Boundary conditions
left_x = 0
right_x = 1


xs = np.arange(0,1+hx,hx)

ts = np.arange(0,tf+ht, ht)

Us = [[]]

## setting initial condition on t = 0
for x in xs:
    if (x < 0.5):
        newU = 0
    else:
        newU = 1
    Us[-1].append(newU)
        
        
## solving for t+1

for t in ts[:-1]:
    U = []
    for x in range(1,len(xs)-1):   
        newU = (1-sigma)*Us[-1][x] + sigma * Us[-1][x-1]
        U.append(newU)
    U.append(1)
    U.insert(0,0)
    Us.append(U)
```

```python
plt.figure(figsize=(8,4))
for i in range(len(Us)):
    plt.plot(xs,Us[i])
plt.grid(alpha = 0.5)
plt.show()
```
