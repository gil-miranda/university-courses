```python
##### Vectorized forward euler
### Input:  F -> Differential equation;
###         y0 -> list or scalar for initial condition;
###         ts -> list of points in time to evaluate the equation;
###         p -> list or scalar for parameters for F, default is set to 0 if F has no extra parameters;
### Output: ys -> numpy array with all solutions for each step t, ys is a Matrix
##### Gil Miranda - last revision 15/09/2019
def f_euler(F, ts, y0, p = 0):
    ys = [y0]
    h = ts[1]-ts[0]
    for tnext in ts[1:]:
        ynext = ys[-1] + F(tnext, ys[-1], p)*h
        ys.append(ynext)
        t = tnext
    return np.array(ys)

```

```python
def contour_up(x):
    return 0

def contour_down(x):
    return 0

def contour_left(x):
    return 500

def contour_right(x):
    return 293

def alpha(x,y):
    return 1.9e-6

def m_Tnew(n, i, Ts):
    if i-1 == 0:
        newT = Ts[i] + k * dt/h**2 * (Ts[i+1] - 2*Ts[i] + alpha(i))
    elif i+1 == N+1:
        newT = Ts[i] + k * dt/h**2 * (beta(i) - 2*Ts[i] + Ts[i-1])
    else:
        newT = Ts[i] + k * dt/h**2 * (Ts[i+1] - 2*Ts[i] + Ts[i-1])
    return newT

Nx = 50
Ny = 52
Nt = 10
ht = 6e-4
hx = 1/(Nx+1)
hy = 1/(Ny-1)

xs = [hx*i for i in range(Nx+2)]
ys = [hy*(j-1) for j in range(Ny+2)]

M = []
Ts = [0]*Nx*Ny
for n in range(0,Nt):
    
    for i in range(0,Nx):
        for j in range(0,Ny):
            ## reenumerating
            nij = (i-1) + (j-1)*Nx
            xi = xs[i]
            yi = ys[i]
            rho_x = ht/hx**2
            rho_y = ht/hy**2
            
            Tnew = (1-(4*alpha(xi,yi)))*Ts[-1][n] 
            if(i-1 == 0):
                Tnew += rho_x*alpha(xi,yi)*contour_left(i)
            else: 
                Tnew += rho_x*alpha(xi,yi)*Ts[-1][n-1]
            if (i+1 == Nx):
                Tnew += rho_x*alpha(xi,yi)*contour_right(i)
            else:
                Tnew += rho_x*alpha(xi,yi)*(Ts[-1][n+1])
            
            if(j-1 == 0):
                Tnew += rho_y*alpha(xi,yi)*contour_down(j)
            else:
                Tnew += rho_y*alpha(xi,yi)*(Ts[-1][n-Nx])
            
            if(j+1 == Ny+1):
                Tnew += rho_y*alpha(xi,yi)*contour_up(j)
            else:
                Tnew += rho_y*alpha(xi,yi)*Ts[-1][n+Nx]
            Ts.append(Tnew)
    M.append(Ts)


```
