```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def jogo(n,m):
    rodada = np.random.choice([-1,1], size = (m,2*n))
    return rodada
```

```python
n, m = 100,500
y = np.cumsum(jogo(n,m), axis = 1)
x = np.arange(1,2*n+1)
```

```python
medias = []
for i in range(n):
    medias.append(np.mean(y[i]))
media_geral = np.mean(medias)
std = np.std(y)
```

### Plot para todos os jogos

```python
for i in range(len(y)):
    plt.plot(y[i])
plt.plot(x,[media_geral]*(2*n), '--', color = 'black', linewidth = '2', alpha = .7)
plt.plot(x, [media_geral + std]*(2*n), '--', color = 'black', linewidth = '2', alpha = .7)
plt.plot(x, [media_geral - std]*(2*n), '--', color = 'black', linewidth = '2', alpha = .7)
plt.show()
```

### K, a variavél aleatória para o último empate

```python
k = []
c = []
for i in range(m):
    if 0 in y[i]:
        k.append((len(y[0]) - list(y[i][::-1]).index(0))//2)
        c.append(list(y[i]).count(0))
    else:
        k.append(0)
        c.append(0)
```

```python
plt.hist(k, bins = np.arange(0,n))
plt.show()
```

```python
plt.hist(c, bins = np.arange(0,n))
plt.show()
```

```python
plt.scatter(k,c, marker = 'x', alpha = .7)
plt.title('')
plt.ylabel('a')
plt.xlabel('k')
plt.grid(alpha=.3)
plt.show()
```
