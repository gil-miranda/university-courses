```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rodada(n):
    rodada_i = np.random.randint(2, size = 2*n)
    rodada = [i if i == 1 else -1 for i in rodada_i]
    return rodada
```

```python
def jogo(n, m):
    y = [rodada(n) for i in range(m)]
    return y
```

```python
n, m = 1000,500
y = np.cumsum(jogo(n,m), axis = 1)
x = np.arange(n)
```

```python
plt.plot(x, [0]*n)
for i in range(len(y)):
    plt.plot(y[i])
```

```python
k = []
for i in range(m):
    if 0 in y[i]:
        k.append(list(y[i][::-1]).index(0))
    else:
        k.append(0)
```
