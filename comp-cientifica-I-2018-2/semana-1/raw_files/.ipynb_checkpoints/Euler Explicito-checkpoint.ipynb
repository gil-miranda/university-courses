# Um exemplo de computação científica: Resolvendo EDOs

Imagine que desejamos resolver a seguinte EDO:
$$ \begin{cases} y'(t) = 3y(t) \\ y(0) = 1.2\end{cases} $$
Imagine, também, que não sabemos a solução analítica desta equação.

Como fazer para aproximar a solução?

# Metodologia

1. Modelar o problema (matematizar)
2. Conceber um algoritmo (sucessão de cálculos simples, para resolver um problema maior)
3. Implementar o algoritmo em um programa (traduzir os cálculos matemáticos em informáticos)

## 1. Modelo

O que quer dizer "aproximar a solução"?  O que quer dizer "a solução"?

Uma tentativa de resposta:
- Calcular valores aproximados de $y(t)$ em alguns pontos $t_i$ dados a priori.
- Garantir que o erro de aproximação não seja maior do que $\varepsilon$, dado a priori ou a posteriori.

## 2. Algoritmo

1. Escolher pontos $t_i$ igualmente espaçados num intervalo $[0,T]$
2. Aproximar $y(t_{i+1}) \sim y(t_i) + y'(t_i) \cdot (t_{i+1} - t_i) = y(t_i) + 3y(t_i) \cdot (t_{i+1} - t_i)$
