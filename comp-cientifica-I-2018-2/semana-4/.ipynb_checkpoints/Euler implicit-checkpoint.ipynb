# Uma outra aproximação do método de Euler

Agora, vamos usar $y(t_{i+1}) \sim y(t_i) + y'(t_{i+1}) (t_{i+1} - t_i) = y(t_i) + F(t_{i+1}, y(t_{i+1})) (t_{i+1} - t_i)$.
Só que não podemos calcular $F(t_{i+1}, y(t_{i+1}))$ se não sabemos quanto vale $y(t_{i+1})$.
Portanto, a equação acima é apenas uma fórmula *implícita* para $y(t_{i+1})$...

Em alguns casos, podemos resolver:

```python
def G(t,y):
    return -10*y
```

Se $G(t,y) = -10y$, como da vez passada, temos então
$y_{next} = y_{curr} + [-10(y_{next})]h$, ou seja,
$$ y_{next} (1 + 10h) = y_{curr}, $$
que não apresentará oscilação, para qualquer valor (positivo, claro!) de $h$.

# Passando para uma função

Temos que resolver, a cada etapa, a equação

$$ y(t_{i+1}) = y(t_i) + F(t_{i+1}, y(t_{i+1})) (t_{i+1} - t_i), $$

dados $y(t_i)$, $t_{i+1}$ e $t_i$.

Re-escrevendo em função dos passos:

$$ \delta y = F(t_{i+1}, y + \delta y) h.$$

Assim, precisamos de uma função que retorne $\delta y$, dados $y(t_i)$, $t_{i+1}$ e $h$.

### Nosso caso

$dy = -10(y + dy)h$ sse $dy(1 + 10h) = -10yh$ sse $dy = -10hy/(1 + 10h)$.

# Mais experiências interativas!

```python
from ipywidgets import interact, IntSlider
```

```python
def do(n):
    ts = np.linspace(0,2,num=n)
    ys_i = implicit_euler(solveG, 0, 1, ts[1:])
    ys_e = euler_npts(G, [0,2], 1, npts=n-1)

    fig, axs = plt.subplots(ncols=2, figsize=(13,4))
    
    for ax in axs:
        ax.plot(ts, ys_i, 'o:', label='euler implicit')
        ax.plot(ts, ys_e, 'x-', label='euler explicit')
        ax.plot(ts, np.exp(-10*ts), label='analitic')
        ax.legend()
        ax.set_title("Solução")
    
    axs[1].set_yscale('log')
    plt.show()
```

```python
ns = IntSlider(value=14, min=5, max=200, step=1)
interact(do, n=ns);
```
