```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Metódo de Newton
def newton(f,df,x, prec=1e-8, maxiter=100, iterr = 0, lst = [], reti = False):
    if maxiter == iterr:
        return None
    dx = f(x)/df(x)
    newx = x - dx
    xs = lst[:]
    xs.append(x)
    if abs(dx) < prec:
        if reti:
            return xs, iterr
        else:
            return xs
    else:
        return newton(f,df,newx, prec,iterr = iterr+1, lst = xs)
```

### Questão 1

O próximo ponto do metódo de Newton de uma função $f$ é dado por:
$$
\begin{align}
x_{f_{n+1}} = x_n - \frac{f(x)}{f'(x)} &&(1)
\end{align}
$$
Como: $g(x) = -f(x)$ <br>
Teremos: $g'(x) = -f'(x)$ <br>
Olhando agora para o próximo ponto do metódo de Newton na função $g$ <br>
$$
\begin{align}
x_{g_{n+1}} = x_n - \frac{g(x)}{g'(x)} \implies x_{g_{n+1}} = x_n - \frac{-f(x)}{-f'(x)} \implies x_{g_{n+1}} = x_n - \frac{f(x)}{f'(x)} &&(2)
\end{align}
$$
Temos então (1) = (2), e portanto o metódo de Newton produz os mesmo pontos para $g$ e para $f$<br>
Como os pontos gerados são os mesmos para $f(x)$ e para $-f(x)$<br>
podemos sempre supor $f '(z) > 0$, pois caso $f ' (z) < 0$, basta tomar $-f ' (z)$, pois: <br>
$f'(z) < 0 \implies -f'(z) > 0$

Os pontos de $g(x)$ são dados por:<br>
$$\begin{align}
g_{next}(x) &= x - \frac{g(x)}{g'(x)} \\
&= x - \frac{f(x+b)}{f'(x+b)}
\end{align}
$$
E os pontos de $ f(x)$ são dados por:
$$
f_{next}(x) = x - \frac{f(x)}{f'(x)}\\
$$
Tomando o ponto transladado $x-b$
$$
\begin{align}
g_{next}(x-b) &= (x-b) - \frac{f(x-b+b)}{f'(x-b+b)} \\
&= x - b - \frac{f(x)}{f'(x)} \\
&= f_{next}(x) - b\\
\end{align}
$$
> Que nada mais são do que os pontos iterados em $ f $ e transladados por $ b$

$$

\frac{1}{2} < f'(x) < \frac{3}{2}

Analogamente para $f''(x)$ chegamos a mesma conclusão
$$
\frac{1}{2} < f''(x) < \frac{3}{2} \\
