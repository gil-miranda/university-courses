```python
mt = 5.9742e24
ms = 1.98892e30
G = 6.67428e-11
pos_x = [149.6e9]
pos_y = [0]
v_x = [0]
v_y = [3e3]
s_x = 0
s_y = 0
```

```python
def acc(x,y):
    return [-G*ms*x/(((x-s_x)**2+(y-s_y)**2)**(1/2))**3,-G*ms*y/(((x-s_x)**2+(y-s_y)**2)**(1/2))**3]
```

```python
(((pos_x[-1]-s_x)**2)**(1/2))**3,-G*ms*pos_x[-1]
```

```python
h = 0.1
ts = np.arange(0,3000)
for t in ts:
    nx,ny = acc(pos_x[-1],pos_y[-1])
    pos_x.append(pos_x[-1] + h*v_x[-1])
    pos_y.append(pos_y[-1] + h*v_y[-1])
    
    v_x.append(v_x[-1] + h*nx)
    v_y.append(v_y[-1] + h*ny)
```

```python
plt.plot(pos_x,pos_y)
plt.show()
```
