# Modelo Uniforme Discreto

Seja $X$ uma variável discreta cujos resultados dos eventos sejam equiprováveis, dizemos que $X$ tem uma distribuição uniforme e usamos a notação $X \tilde U_d$
Ou seja, sejam os eventos
$$
x_1, x_2, \dots, x_n
$$
com função de probabilidade dada por
$$
p(x_i) = \frac{1}{n}, \, i \in 1,2,\dots, n
$$

```python
def distUniform(x):
    n = len(x)
    return [1/n for i in x]
```

```python
x = np.linspace(0,50, 51)
```

```python
y = distUniform(x)
```

```python
plt.figure(figsize= (20,10))
plt.bar(x,y)
plt.title('Distribuição Uniforme para uma variável aleatória $X$ que assume valores entre 0 e 50')
plt.show
```

## Esperança da distribuição
Seja $n$ a quantidade de resultados do experimento, com seus resultados discretamente distribuidos em $[a,b]$, com $a,b$ fazendo parte e sendo respectivamente o menor e maior valor, temos:
$$
\begin{align}
\mathbb{E}(X) &= \sum_{i=1}^{n} x_i p(x_i)\\
&= \sum_{i=1}^{n} x_i \frac{1}{n}\\
&= \frac{1}{n} \sum_{i=1}^{n} x_i\\
&= \frac{1}{n} \frac{n(a+b)}{2}\\
&= \frac{a+b}{2}
\end{align}
$$
